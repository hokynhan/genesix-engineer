```python
import numpy as np
import pandas as pd
from pmdarima import auto_arima
import random

values = [40,48,40,54,73,81,82,109,115,117,134,138,151,151,175,186,186,206,213,213,232,235,254,254]

# Generate a list of dates starting from the end of a month and going back in time
dates = pd.date_range(end='2026-03-31', periods=24, freq='ME')

# Create a Pandas dataframe with the dates and values
df = pd.DataFrame({'Date': dates, 'Value': values})

# Set the 'Date' column as the index of the dataframe
df = df.set_index('Date')

# Split the data into training and test sets
# Forecast next 3 months --> Test data should be 3 months
# Forecast next 6 months --> Test data should be 6 months
train = df[:-3]
test = df[-3:]
```

```python
#Fit an arima model to the training data
arima_model = auto_arima(train, seasonal=False, stepwise=False)
# sarima_model = auto_arima(train, seasonal=True, m=12) --> seasonal period to 12 since we'are using monthly data
print(arima_model.order) #(p,d,q)
```

```python
from sklearn.metrics import mean_squared_error

# Make predictions on the test data
predictions = arima_model.predict(n_periods=len(test))

# Calculate the mean squared error
mse = mean_squared_error(test, predictions)

# Print the mean squared error
print("Mean Squared Error:", mse)
```
