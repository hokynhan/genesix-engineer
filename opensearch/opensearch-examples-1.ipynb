```python
from opensearchpy import OpenSearch

os_client = OpenSearch(hosts=["http://admin:UOUnEx4pro92mhQz@localhost:9200"])

print(os_client.info())
```

```python
# Exercise 1 - Production catalog - Create index
index_name = "products"

mapping = {
    "mappings": {
        "properties": {
            "product_id": {"type": "keyword"},
            "name": {"type": "text"},
            "category": {"type": "keyword"},
            "price": {"type": "float"},
            "created_at": {"type": "date"}
        }
    }
}

# Create index if not exists
if not os_client.indices.exists(index=index_name):
    os_client.indices.create(index=index_name, body=mapping)
    print("Products index created")
```

```python
# Exercise 1 - Production catalog Insert sample data
products = [
    {"product_id":"P001","name":"Apple MacBook Air M1","category":"laptop","price":999,"created_at":"2026-03-01"},
    {"product_id":"P002","name":"Dell XPS 13","category":"laptop","price":1200,"created_at":"2026-03-02"},
    {"product_id":"P003","name":"Sony WH1000XM5 Headphones","category":"audio","price":399,"created_at":"2026-03-03"}
]

for p in products:
    os_client.index(index="products", body=p)

print("Inserted product data")
```

```python
# Exercise 1 - Production catalog - Search products
import json
query = {
    "query": {
        "match": {
            "name": "macbook"
        }
    }
}

response = os_client.search(index="products", body=query)

print(json.dumps(response["hits"]["hits"], indent=2))
```

```python
#Exercise 1 - Delete an index

index_name = "products"

if os_client.indices.exists(index=index_name):
    os_client.indices.delete(index=index_name)
    print(f"Index '{index_name}' deleted")
else:
    print(f"Index '{index_name}' deos not exist")
```
